# **2. Data Cleaning: (Score : 2)**
Find the missing and inappropriate values, treat them appropriately.
Remove all duplicate rows.
Find the outliers.
Replace the value 0 in age as NaN
Treat the null values in all columns using any measures(removing/ replace the values with mean/median/mode)


```python
#Load a dataset
df = pd.read_csv('/content/Employee (1).csv')
```

```python
# Check for missing values in each column
missing_values = df.isnull().sum()
print("Missing Values:")
print(missing_values)
```

```python
# Check for inappropriate values (e.g., age = 0)
inappropriate_age = df[df['Age'] == 0]
print("Rows with Age = 0:")
print(inappropriate_age)
```

```python
# Remove duplicate rows
df = df.drop_duplicates()

# Verify the changes
print(f"Number of rows after removing duplicates: {len(df)}")
```

```python
# Function to detect outliers using IQR
def find_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return column[(column < lower_bound) | (column > upper_bound)]

# Find outliers in numerical columns
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
outliers = {col: find_outliers(df[col]) for col in numerical_columns}

# Print outliers
for col, outlier_values in outliers.items():
    print(f"Outliers in {col}:")
    print(outlier_values)
    print("-" * 40)
```

```python
import numpy as np
```

```python
# Replace 0 in 'age' with NaN
df['Age'] = df['Age'].replace(0, np.nan)

# Verify the changes
print("Rows with Age as NaN:")
print(df[df['Age'].isnull()])
```

```python


# Treat null values
for col in numerical_columns:
    df[col].fillna(df[col].median(), inplace=True)

categorical_columns = df.select_dtypes(include=['object']).columns
for col in categorical_columns:
    df[col].fillna(df[col].mode()[0], inplace=True)

print("Null values after treatment:")
print(df.isnull().sum())

# Save cleaned dataset
df.to_csv('cleaned_dataset.csv', index=False)
```

# **3. Data Analysis: (Score : 2)**
Filter the data with age >40 and salary<5000
Plot the chart with age and salary
Count the number of people from each place and represent it visually


```python

import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('/content/Employee (1).csv')

# Filter the data
filtered_data = df[(df['Age'] > 40) & (df['Salary'] < 5000)]
print("Filtered Data (Age > 40 and salary < 5000):")
print(filtered_data)

# Plot a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(filtered_data['Age'], filtered_data['Salary'], color='blue', alpha=0.5)
plt.title('Age vs Salary (Age > 40 and Salary < 5000)')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)
plt.show()

# Count the number of people from each place
place_counts = df['Place'].value_counts()
print("Number of people from each place:")
print(place_counts)

# Plot a bar chart
plt.figure(figsize=(10, 6))
place_counts.plot(kind='bar', color='green')
plt.title('Number of People from Each Place')
plt.xlabel('Place')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

# **4. Data Encoding: (Score : 2)**
Convert categorical variables into numerical representations using techniques such as one-hot encoding, label encoding, making them suitable for analysis by machine learning algorithms.


```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# Load dataset
df = pd.read_csv('/content/Employee (1).csv')

# Identify categorical columns
categorical_columns = df.select_dtypes(include=['object', 'category']).columns
print("Categorical Columns:")
print(categorical_columns)

# One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=categorical_columns, drop_first=True)
print("DataFrame after One-Hot Encoding:")
print(df_encoded.head())

# Label Encoding (alternative)
label_encoder = LabelEncoder()
for col in categorical_columns:
    df[col] = label_encoder.fit_transform(df[col])
print("DataFrame after Label Encoding:")
print(df.head())

# Save encoded dataset
df_encoded.to_csv('encoded_dataset.csv', index=False)
```

# **5. Feature Scaling: (Score : 2)**
After the process of encoding, perform the scaling of the features using standardscaler and minmaxscaler.


```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Load dataset
df = pd.read_csv('/content/Employee (1).csv')

# Check column names
print("Columns in the dataset:")
print(df.columns)

# If 'City' exists, perform One-Hot Encoding
if 'City' in df.columns:
    df_encoded = pd.get_dummies(df, columns=['City'], drop_first=True)
else:
    print("Column 'City' not found. Skipping One-Hot Encoding.")
    df_encoded = df

# Select numerical columns
numerical_columns = df_encoded.select_dtypes(include=['int64', 'float64']).columns
print("Numerical Columns:")
print(numerical_columns)

# Standard Scaling
standard_scaler = StandardScaler()
df_standardized = df_encoded.copy()
df_standardized[numerical_columns] = standard_scaler.fit_transform(df_encoded[numerical_columns])
print("DataFrame after Standard Scaling:")
print(df_standardized.head())

# MinMax Scaling
minmax_scaler = MinMaxScaler()
df_minmax_scaled = df_encoded.copy()
df_minmax_scaled[numerical_columns] = minmax_scaler.fit_transform(df_encoded[numerical_columns])
print("DataFrame after MinMax Scaling:")
print(df_minmax_scaled.head())

# Save scaled datasets
df_standardized.to_csv('standardized_dataset.csv', index=False)
df_minmax_scaled.to_csv('minmax_scaled_dataset.csv', index=False)
```
